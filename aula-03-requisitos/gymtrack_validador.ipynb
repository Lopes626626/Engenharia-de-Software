```python
import time

print("🏋️ GymTrack – Validador de Treino")
print("=" * 40)

# --- DADOS DO TREINO ---
exercicio = "Abdominais"
peso_kg   = 80
repeticoes = 10

# -----------------------------------------------------------------
# -----------------------------------------------------------------

if exercicio != "":
    print("✅ [RF01] Exercício válido:",exercicio)
else:
    print("❌ [RF01] Erro: O nome do exercício não pode estar vazio.")

# -----------------------------------------------------------------
# -----------------------------------------------------------------

if 1 <= peso_kg <= 300:
    print("✅ [RF02] Peso dentro do limite.")
else:
    print(f"❌ [RF02] Erro: Peso {peso_kg}kg está fora do intervalo permitido (1-300kg).")

# -----------------------------------------------------------------
# -----------------------------------------------------------------
if 1 <= repeticoes <= 50:
    print("✅ [RF03] Repetições válidas.")
else:
    print(f"❌ [RF03] Erro: {repeticoes} repetições é um valor inválido (1-50).")

# -----------------------------------------------------------------
# -----------------------------------------------------------------

print("=" * 40)

inicio = time.time()

time.sleep(0.05)

print(f"✅ Série registrada: {exercicio} | {peso_kg}kg x {repeticoes} reps")

fim = time.time()
tempo_ms = (fim - inicio) * 1000

if tempo_ms < 200:
    print(f"✅ [RNF01] Tempo de registro: {tempo_ms:.0f}ms <- dentro do limite!")
else:
    print(f"❌ [RNF01] Lento demais: {tempo_ms:.0f}ms <- limite é 200ms")

# REFLEXÃO
#1. O RF foca na ação de negócio, enquanto o RNF foca na qualidade e restrição técnica
#2. Sem o RNF, o usuário ficaria frustado com a lentidão do app no meio do treino, podendo até desistir de registrar as séries.
#3. RNF Adicional: Seguraça (O sistema deve exigir autenticação biométrica ou senha para acessar o perfil do aluno).
```
