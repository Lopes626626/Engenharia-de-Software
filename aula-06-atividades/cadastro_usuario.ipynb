```python
def processo_login_google(usuario_autorizou: bool, token_valido: bool) -> str:
    """
    Simula o fluxo do Diagrama de Atividades de Login com Google.
    """
    print("\n▶ [INÍCIO] Iniciando processo de login...")
    print("  → Redirecionando para Google Auth")


    if not usuario_autorizou:
        print("  ✗ [DECISÃO] Usuário NÃO autorizou")
        print("  → Redirecionando com erro")
        return "🛑 ERRO: Acesso negado pelo usuário\n"

    print("  ✓ [DECISÃO] Usuário autorizou")
    print("  → Token de acesso recebido do Google")


    if not token_valido:
        print("  ✗ [DECISÃO] Token INVÁLIDO")
        print("  → Exibindo erro de autenticação")
        return "🛑 ERRO: Token inválido ou expirado\n"

    print("  ✓ [DECISÃO] Token VÁLIDO")
    print("  → Sessão criada no sistema")
    print("  → Redirecionando ao dashboard")
    print("⊗ Processo concluído")
    return "✅ SUCESSO: Login realizado!\n"


# Executando os 3 caminhos possíveis do diagrama:
print("=== TESTE 1: Caminho Feliz (Autorizou + Token Válido) ===")
print(processo_login_google(True, True))

print("=== TESTE 2: Caminho Alternativo (Autorizou + Token Inválido) ===")
print(processo_login_google(True, False))

print("=== TESTE 3: Caminho Alternativo (Não autorizou) ===")
print(processo_login_google(False, True))
```
