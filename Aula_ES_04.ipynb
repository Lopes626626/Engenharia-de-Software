```python
# ============================================================
# 🚀 SRS em Python — Rode no Google Colab!
# Aula 04 — Engenharia de Software · FIAP
# ============================================================
from dataclasses import dataclass, field
from typing import List
from enum import Enum

class Prioridade(Enum):
    ALTA = "Alta"
    MEDIA = "Média"
    BAIXA = "Baixa"

@dataclass
class RequisitoFuncional:
    id: str
    nome: str
    descricao: str
    prioridade: Prioridade
    ator: str
    pre_condicao: str
    pos_condicao: str

@dataclass
class RequisitoNaoFuncional:
    id: str
    categoria: str
    descricao: str
    criterio_aceitacao: str

@dataclass
class SRS:
    projeto: str
    versao: str
    descricao: str
    requisitos_funcionais: List[RequisitoFuncional] = field(default_factory=list)
    requisitos_nao_funcionais: List[RequisitoNaoFuncional] = field(default_factory=list)

    def adicionar_rf(self, req: RequisitoFuncional):
        self.requisitos_funcionais.append(req)
        print(f"✅ RF '{req.id}' adicionado!")

    def adicionar_rnf(self, req: RequisitoNaoFuncional):
        self.requisitos_nao_funcionais.append(req)
        print(f"✅ RNF '{req.id}' adicionado!")

    def relatorio(self):
        print(f"\n{'='*50}")
        print(f"📋 SRS — {self.projeto} v{self.versao}")
        print(f"{'='*50}")
        print(f"📝 {self.descricao}\n")

        print(f"🔧 REQUISITOS FUNCIONAIS ({len(self.requisitos_funcionais)})")
        for rf in self.requisitos_funcionais:
            print(f"  [{rf.id}] {rf.nome} — Prioridade: {rf.prioridade.value}")
            print(f"       Ator: {rf.ator}")
            print(f"       📌 {rf.descricao}\n")

        print(f"⚡ REQUISITOS NÃO-FUNCIONAIS ({len(self.requisitos_nao_funcionais)})")
        for rnf in self.requisitos_nao_funcionais:
            print(f"  [{rnf.id}] {rnf.categoria}")
            print(f"       📌 {rnf.descricao}")
            print(f"       ✔️  Critério: {rnf.criterio_aceitacao}\n")

def validar_requisito(rf: RequisitoFuncional) -> dict:
    """
    Valida se um requisito funcional segue as boas práticas.
    Retorna um dict com os resultados da validação.
    """
    resultados = {

        "descricao_valida": len(rf.descricao) > 20,


        "pre_condicao_valida": rf.pre_condicao.strip() != "",


        "criterio_mensuravel": any(char.isdigit() for char in rf.descricao)
    }

    return resultados

# ---- Criando o SRS do App de Delivery ----
srs = SRS(
    projeto="Sistema de Pedidos (FIAP Delivery)",
    versao="1.0",
    descricao="Plataforma para gestão e rastreamento de pedidos."
)

rf1 = RequisitoFuncional(
    id="RF-001",
    nome="Realizar Pedido",
    descricao="O sistema deve permitir que o cliente selecione itens do cardápio de um restaurante, adicione ao carrinho e confirme o pedido com endereço de entrega.",
    prioridade=Prioridade.ALTA,
    ator="Cliente",
    pre_condicao="Cliente autenticado, restaurante com status 'Aberto'",
    pos_condicao="Pedido criado com status 'Aguardando confirmação', restaurante notificado"
)
srs.adicionar_rf(rf1)

rf2 = RequisitoFuncional(
    id="RF-002",
    nome="Confirmação de Pedido pelo Restaurante",
    descricao="O sistema deve exibir ao restaurante os pedidos recebidos em tempo real e permitir confirmação ou rejeição com motivo.",
    prioridade=Prioridade.ALTA,
    ator="Restaurante",
    pre_condicao="Pedido com status 'Aguardando confirmação'",
    pos_condicao="Status atualizado para 'Em preparo' ou 'Cancelado'"
)
srs.adicionar_rf(rf2)

rf3 = RequisitoFuncional(
    id="RF-003",
    nome="Rastreamento do Entregador em Tempo Real",
    descricao="O sistema deve exibir a posição GPS do entregador no mapa, atualizada a cada 3 segundos, enquanto o pedido estiver com status 'Em rota'.",
    prioridade=Prioridade.ALTA,
    ator="Cliente",
    pre_condicao="Pedido com status 'Em rota', entregador com GPS ativo e app aberto",
    pos_condicao="Cliente visualiza localização atual e ETA atualizado do entregador"
)
srs.adicionar_rf(rf3)


srs.adicionar_rnf(RequisitoNaoFuncional(
    id="RNF-001",
    categoria="Disponibilidade",
    descricao="O sistema deve ter uma disponibilidade de 99.9% durante o horário comercial.",
    criterio_aceitacao="Relatórios mensais de uptime do servidor AWS não apontarem mais de 43 minutos de queda."
))

srs.adicionar_rnf(RequisitoNaoFuncional(
    id="RNF-002",
    categoria="Conformidade (LGPD)",
    descricao="Os dados pessoais dos clientes, como endereço e nome, devem ser anonimizados no banco de dados para analistas.",
    criterio_aceitacao="Aprovação em auditoria de segurança de dados simulando vazamento."
))

srs.relatorio()

print("\n" + "="*50)
print("🔍 TESTANDO A VALIDAÇÃO DOS REQUISITOS")
print("="*50)

for requisito in srs.requisitos_funcionais:
    resultado_validacao = validar_requisito(requisito)
    print(f"\nAnalisando {requisito.id}:")
    print(f" - Descrição > 20 caracteres: {'✅' if resultado_validacao['descricao_valida'] else '❌'}")
    print(f" - Tem pré-condição: {'✅' if resultado_validacao['pre_condicao_valida'] else '❌'}")
    print(f" - Tem critério mensurável (números na desc.): {'✅' if resultado_validacao['criterio_mensuravel'] else '❌'}")
```
