```python
# --- DEFINIÇÃO DAS CLASSES ---

class Filme:
    def __init__(self, titulo: str, duracao: int, genero: str):
        self.titulo = titulo
        self.duracao = duracao
        self.genero = genero

class Avaliacao:
    def __init__(self, nota: float, comentario: str):
        self.nota = nota
        self.comentario = comentario
        self.filme = None

class Catalogo:
    def __init__(self, titulo: str, qtdFilmes: int):
        self.titulo = titulo
        self.qtdFilmes = qtdFilmes
        self.filmes: list[Filme] = []

    def add_filme(self, filme: Filme):
        self.filmes.append(filme)
        self.qtdFilmes = len(self.filmes)

    def listar_filmes(self):
        print(f"\n--- Catálogo: {self.titulo} ---")
        for f in self.filmes:
            print(f"🎬 {f.titulo} ({f.duracao} min) - {f.genero}")

class Plataforma:
    def __init__(self, nome: str, pais: str):
        self.nome = nome
        self.pais = pais


class Usuario:
    def __init__(self, nome: str, email: str, plano: str):
        self.nome = nome
        self.email = email
        self.plano = plano
        self.avaliacoes: list[Avaliacao] = []

    def avaliar(self, filme: Filme, avaliacao: Avaliacao):
        avaliacao.filme = filme
        self.avaliacoes.append(avaliacao)
        print(f"✅ {self.nome} avaliou o filme '{filme.titulo}' com sucesso!")

    def ver_avaliacoes(self):
        print(f"\n--- Avaliações de {self.nome} ---")
        for a in self.avaliacoes:

            nome_filme = a.filme.titulo if a.filme else "Filme desconhecido"
            print(f"⭐ {a.nota}/10 para '{nome_filme}': \"{a.comentario}\"")


# --- TESTE ---

netflix = Plataforma("Netflix", "EUA")
catalogo = Catalogo("Filmes em Destaque", 0)

filme1 = Filme("Oppenheimer", 180, "Drama")
filme2 = Filme("Barbie", 114, "Comédia")

catalogo.add_filme(filme1)
catalogo.add_filme(filme2)

usuario = Usuario("Ana", "ana@email.com", "Premium")
avaliacao = Avaliacao(9.5, "Incrível! Assisti duas vezes")

# Relacionando os objetos
usuario.avaliar(filme1, avaliacao)

# Exibindo os resultados
catalogo.listar_filmes()
usuario.ver_avaliacoes()
```
